1. Ask the user for a string that contains both letters and numbers (e.g., "Py2th0on3"). Write a
program that separates them into two lists: one for the letters and one for the numbers. Calculate the
sum of the numbers found and join the letters to form a word.


```python
text = input("write a string that contains both letters and numbers: ")
sum = 0
letters = []

for i in text:
    if i.isdigit():
        sum += int(i)
    else:
        letters.append(i)

word = "".join(letters)
print(f'word {word}, sum : {sum}')
```

3. A store gives discounts based on the total price: Under $50 (No discount), $50 to $100 (10%
discount), and Over $100 (20% discount). If the user is a "Member," they get an additional 5% off the
final calculated price.

```python
def calculate_discount(price, isMember):
  discount = 0
  if price >= 50 and price <= 100:
    discount = 0.1 * price
  elif price > 100:
    discount = 0.2 * price
    # calculate total price after discount
  totalPrice = price - discount
  # if member, apply additional 5% discount
  if isMember == "y":
    totalPrice -= 0.05 * totalPrice
  return totalPrice

calculate_discount( int(input("Enter the total price: ")),input("Are you a member? (y/n): "))
```

4.Ask the user to input two separate lists of names (comma-separated). Create a third list that
contains only the names that appear in both lists

```python
def common_names(list1, list2):
  """
  This function takes two comma-separated lists of names and returns a third list containing only the names that appear in both input lists.
  """
  list3 = []
  list1 = list1.split(",")
  list2 = list2.split(",")
  # convert lists to sets
  set1 = set(list1)
  set2 = set(list2)
  # find intersection of sets
  intersection = set1 & set2
  # convert intersection back to list
  list3 = list(intersection)
  print(f"common names: {list3}")
# call the function
common_names(input("Enter the first list of names: "), input("Enter the second list of names: "))
```

7.Ask for a multi-digit number. Sum its digits. If the result is more than one digit, keep summing the
digits until you get a single-digit result.

```python
def multiDigitNumber(num):
    """
    Ask for a multi-digit number. Sum its digits. 
    If the result is more than one digit, keep summing the digits
    until you get a single-digit result
    """
    
    # conver number to list of numbers and convert to number
    bego = [int(x) for x in str(num)]
    
    # if result more that one
    while len(bego) > 1:
        total = sum(bego)
        # convert result to list of numbers
        bego = [int(x) for x in str(total)]
    return bego[0]

# قراءة الرقم من المستخدم وتحويله لـ int
num = int(input("Please enter a multi-digit number: "))
result = multiDigitNumber(num)
print("Digital Root:", result)
```

```python
def repeatedNumber(numbersList):
    # make list of uniqu numbers because it will make map on it so don't repeated tuple on map 
    uniquNumbes = set(numbersList)
    # map will return as tuple then will put in array
    result = list(map(lambda x: (x, numbersList.count(x)), uniquNumbes)) 
    return result



listString = list(input("please number that will be list "))
numbers = [int(x) for x in listString]

repeatedNumber(numbers)
```

11. Check if a full sentence is a palindrome, ignoring spaces and capitalization

```python
def palinrome(sentence):
    """
    Check if a full sentence is a palindrome, ignoring spaces and capitalization
    """
    # remove spaces and convert to lowercase
    cleaned_sentence = ''.join(sentence.split()).lower().strip()
    # check if cleaned sentence is equal to its reverse
    if cleaned_sentence == cleaned_sentence[::-1]:
        print("The sentence is a palindrome.")
        return True
    else:
        print("The sentence is not a palindrome.")
        return False
    
input_sentence = input("Enter a sentence: ")
palinrome(input_sentence)
```

13.In a list of numbers, find all "peaks." A peak is a number that is greater than the number to its left
and the number to its right.

```python
def list_peaks(listNumbers):
    print(listNumbers)
    newList = [
        elem
        for index, elem in enumerate(listNumbers)
        if (
            index > 0
            and index < len(listNumbers) - 1
            and listNumbers[index] > listNumbers[index - 1]
            and listNumbers[index] > listNumbers[index + 1]
        )
    ]
    print(newList)
    return newList


InputPeaks = list(input("please enter list of numbers separated by commas: ").split(","))
list_peaks(InputPeaks)
```

15. Create a dictionary where the keys are the vowels found in a user's string and the values are lists
of the indices where those vowels appear.
• Example Input: "banana"
• Example Output: {'a': [1, 3, 5]}

```python
def vowels_indices(text):
    vowels = "aeiou"
    result = {}

    for index, char in enumerate(text):
        if char in vowels:
            if char not in result:
                result[char] = []
            result[char].append(index)

    return result


text = input("Enter a string: ")
print(vowels_indices(text))
```
