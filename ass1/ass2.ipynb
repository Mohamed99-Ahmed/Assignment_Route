## Write a function that takes a list of strings and a number K. The function should
   return a new list containing only the strings that have a length greater than K.
   • Input: list=["apple", "pie", "banana"], K=3
   • Output: ['apple', 'banana']

```python
def list_arrs(list , k) : 
    newList = [i for i in list if len(i) > k]
    return newList 

print(list_arrs(["apple", "pie", "banana"], 5))  # Expected output: ['apple', 'banana']
```

## Write a program that takes an email address as input and splits it into two parts: the
"username" and the "domain".
• Example Input: support@google.comExample
• Output: Username: support, Domain: google.com

```python
def seperateEmail(email):
    username, domain = email.split('@')
    return f"username: {username}, domain: {domain}"

print(seperateEmail("mohamed@company.com"))
```

## Write a program to find the sum of all numbers in a list. To make it mid-level, ensure
the program ignores any elements in the list that are not numbers (like strings or
booleans).
• Example Input: [10, "apple", 20, True, 5]
• Example Output: Total Sum: 35

```python
def sumNumbesOfList(numbers):
    newList = [i for i in numbers if isinstance(i, (int, float)) ]
    return sum(newList)

print(sumNumbesOfList([1, 2, '3', 4.5, 'hello', 6])) 
```

## 6. Write a Python class Restaurant with attributes like menu_items, book_table, and
customer_orders, and methods like add_item_to_menu, book_tables, and
customer_order. (Note: Use dictionaries and lists to store the data.)

```python
from abc import ABC as abs, abstractmethod
class Resturant(abs): 
    def __init__(self):
        self.menus = []
        self.book_table = []
        self.customer_orders = {}
    # it is must add a menu to the menus list
    @abstractmethod
    def add_menu(self, menuList: list):
        self.menus = menuList
        return self.menus
    def book_tables(self, tableNumber:int):
        if tableNumber in self.book_table:
            return "Table already booked"
        self.book_table.append(tableNumber)
        return self.book_table
    # check order value are in menu list
    def check_order_in_menu(self, order: dict):
        for item in order.keys():
            if item not in self.menus:
                return False
        return True
    def add_customer_order(self, order: dict, tableNumber:int):
        if not self.check_order_in_menu(order):
            return "Order contains items not in the menu"
        self.customer_orders[tableNumber] = order
        if tableNumber  in self.book_table:
           final_order = {tableNumber: order}
           return final_order
        else: 
            return "Table not booked"
    
class MyResturant(Resturant):
    def add_menu(self, menuList: list):
        super().add_menu(menuList)
        return f"Menu added: {self.menus}"
    
my_resturant = MyResturant()
# add menu items
my_resturant.add_menu(['Pizza', 'Pasta', 'Salad'])
#  Make table reservations.
my_resturant.book_tables(3)
my_resturant.book_tables(4)
my_resturant.book_tables(5)
# Add customer orders
print(my_resturant.add_customer_order({'Pizza': 2, 'sfd': 1}, 3))
print(my_resturant.add_customer_order({'Pizza': 2, 'Salad': 1}, 4))
print(my_resturant.add_customer_order({'Pasta': 1}, 5))

# print orders 
print(f"my orders are: {my_resturant.customer_orders}")
# print tables 
print(f"booked tables are: {my_resturant.book_table}")




```
