### 2. Handling Missing Values
- Fill the missing Salary values with the median salary of the entire
company.
-  Fill the missing Bonus values with 0.

```python
df["Salary"].fillna(df["Salary"].median())
df["Bonus"].fillna(0)

df.isnull().sum()
```

### Data Cleaning: The Employee_Name column is in the format "FirstName
LastName". Split this column into two new columns: First_Name and Last_Name,
then delete the original Employee_Name column.

```python
# Split the Employee_Name column into First_Name and Last_Name
df[['First_Name', 'Last_Name']] = df['Employee_Name'].str.split(' ', expand=True)

# Delete the original column
df.drop(columns=['Employee_Name'], inplace=True)
```

```python
df.head()
```

### filtering :Create a new DataFrame called Senior_IT that contains only employees
from the 'IT' department who have more than 8 years of experience.


```python
Senior_IT = df.query('Department == "IT" and Salary > 70000')
print(Senior_IT)
```

### 5. Performance Mapping Create a new column called Performance_Level based on
the Evaluation_Score (1-100):
• Score > 80: "High"
• Score 50-80: "Medium"
• Score < 50: "Low"

```python
df["Performance_Level"] = [
    "High" if x > 80 else "Medium" if x >= 50 else "Low"
    for x in df["Evaluation_Score"]
]
```

### . Calculate the total salary expense and the average evaluation score for each
Department.


```python
# calculate the totle salery
totlal_slary = df.groupby("Department")["Salary"].sum()
print(totlal_slary)

# average evualution
average = df.groupby("Department")["Evaluation_Score"].mean()
print(average)
```
